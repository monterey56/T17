```python
# Import pandas
import pandas as pd
```

```python
# Create a DataFrame with balance.txt
df = pd.read_csv('balance.txt',sep=' ')
```

Write the code needed to produce a report that provides the following information:
* Compare the average income based on ethnicity. 
* On average, do married or single people have a higher balance?
* What is the highest income in our dataset?
* What is the lowest income in our dataset?
* How many cards do we have recorded in our dataset? (Hint: use `sum()`)
* How many females do we have information for vs how many males? (Hint: use `count()`For a list of all methods for computation of descriptive stats, explore the [pandas documentation](https://pandas.pydata.org/pandas-docs/stable/reference/frame.html#computations-descriptive-stats)). 




```python
#Compare the average income based on ethnicity
print("The average income of an African American is", df[df.Ethnicity == "African American"].loc[:,"Income"].mean())
print("The average income of an Asian is", df[df.Ethnicity == "Asian"].loc[:,"Income"].mean())
print("The average income of a Caucasian is", df[df.Ethnicity == "Asian"].loc[:,"Income"].mean())
```

```python
#Compare the balance between married and single people
print(df.groupby('Married')['Balance'].mean())
```

```python
#Find the highest and lowest income in the dataset
print("The highest income in the dataset is", df.loc[:, "Income"].max())
print("The lowest income in the dataset is", df.loc[:, "Income"].min())
```

```python
#Find the number of cards recorded in the dataset
print("The total number of cards recorded in the dataset is", df.loc[:, "Cards"].sum())
```

```python
#Find the how many Females vs Males we have recorded data for
females = df[df.Gender == 'Female'].count()
males = df[df.Gender == 'Male'].count()
print("The number of Females we have data recorded for is", df[df.Gender == "Female"].loc[:,"Gender"].count())
print("The number of Males we have data recorded for is", df[df.Gender == "Male"].loc[:,"Gender"].count())
```
